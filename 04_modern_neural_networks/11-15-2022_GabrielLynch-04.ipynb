```python
import numpy as np
import matplotlib.pyplot as plt

outfile = "10115718.output"

with open(outfile) as f:
    lines = f.readlines()
it = []
loss = []
acc = []
for line in lines:
    if line.startswith("Finished step"):
        line_split = line.split(" ")
        it.append(int(line_split[2]))
        loss.append(float(line_split[7].split("=")[-1].replace(",","")))
        acc.append(float(line_split[8].split("=")[-1]))
        
it = np.array(it)
loss = np.array(loss)
acc = np.array(acc)

print(it)
print(loss)
print(acc)
```

```python
fig,ax = plt.subplots(2,1, figsize=(8,10))

ax[0].plot(it, loss, label="loss")
ax[1].plot(it, acc, label="acc")
ax[0].set_xlabel("Iteration")
ax[1].set_xlabel("Iteration")
ax[0].set_ylabel("Loss")
ax[1].set_ylabel("Acc")
ax[0].set_yscale("log")
plt.legend()
plt.savefig("11-15-2022_GabrielLynch-04.pdf",dpi=300, bbox_inches='tight')
```
