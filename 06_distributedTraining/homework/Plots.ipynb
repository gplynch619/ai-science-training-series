```python
import numpy as np
import matplotlib.pyplot as plt

nranks = 2

metrics = np.loadtxt("metrics{}.dat".format(nranks))

print(metrics[:,0])
print(metrics[:,2])


fig,ax = plt.subplots()

ax.plot(metrics[:,0], label="Training acc.")
ax.plot(metrics[:,2], label="Validation acc.")
ax.set_xlabel("Epoch")
ax.set_ylabel("Acc")

plt.legend()
plt.savefig("accuracy_{}.pdf".format(nranks),dpi=300, bbox_inches='tight')
```
